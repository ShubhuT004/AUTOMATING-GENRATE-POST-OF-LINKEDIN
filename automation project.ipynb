```python
import feedparser
import requests
from bs4 import BeautifulSoup
from transformers import pipeline
```

```python
def clean_html(text):
    try:
        return BeautifulSoup(text, "html.parser").get_text()
    except:
        return text
def get_ai_news():
    feeds = [
        "https://feeds.feedburner.com/TechCrunch/artificial-intelligence",
        "https://www.analyticsvidhya.com/feed/",
        "https://www.marktechpost.com/feed/",
    ]

    articles = []

    for url in feeds:
        try:
            response = requests.get(url, timeout=8)
            feed = feedparser.parse(response.text)

            for entry in feed.entries[:3]:
                articles.append({
                    "title": entry.title,
                    "link": entry.link,
                    "summary": clean_html(entry.get("summary", "")),
                })

        except Exception as e:
            print(f"‚ùå Error fetching {url}: {e}")

    return articles

summarizer = pipeline(
    "summarization",
    model="sshleifer/distilbart-cnn-12-6"
)

def summarize_text(text):
    text = text.strip()

    if len(text) < 50:
        return text

    try:
        # Wrap input inside list to avoid model_kwargs error
        result = summarizer([text], max_length=70, min_length=25, do_sample=False)
        return result[0]["summary_text"]
    except Exception as e:
        print("‚ùå Summarization error:", e)
        return text

def generate_post(title, summary, link):
    return f"""
üöÄ {title}

{summary}

üëâ Read more: {link}

#AI #ArtificialIntelligence #MachineLearning #DataScience #TechNews
"""

if __name__ == "__main__":
    news = get_ai_news()

    for article in news:
        summary = summarize_text(article["summary"])
        post = generate_post(article["title"], summary, article["link"])

        print(post)
        print("=" * 120)
```
